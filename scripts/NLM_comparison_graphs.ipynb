```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
class GridGraph:
    def __init__(self,nr_run=500):
        self.arr_backtrack_time = []
        self.arr_grid_time = []
        self.nr_run = nr_run
    
    def get_time(self):
        for i in range(self.nr_run):
            self.arr_backtrack_time.append(random.uniform(0.0,1.0))
            self.arr_grid_time.append(random.uniform(0.0,1.0))
            
obj = GridGraph()
obj.get_time()
#print(obj.arr_backtrack_time,len(obj.arr_backtrack_time))
#print(obj.arr_grid_time,len(obj.arr_grid_time))
```

```python
# graphical representation ----

# for x - axis
x_input = []
for i in range(1,501):
    x_input.append(i)


plt.plot(x_input,obj.arr_backtrack_time)
plt.plot(x_input,obj.arr_grid_time)
plt.xlabel("Number Of Grids")
plt.ylabel("Run Time")
plt.title("Run Time: NLM vs Backtracking")
label = ['Backtrack','NLM']
plt.legend(label)
plt.show() 
```
