```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# List of file names
file_names = ['results_max_steps_81.csv', 'results_max_steps_150.csv', 'results_max_steps_400.csv', 'results_max_steps_729.csv',
              'results_max_steps_81_10epochs.csv', 'results_max_steps_150_10epochs.csv', 'results_max_steps_400_10epochs.csv', 'results_max_steps_729_10epochs.csv',
              'results_max_steps_81_100epochs.csv', 'results_max_steps_150_100epochs.csv', 'results_max_steps_400_100epochs.csv', 'results_max_steps_729_100epochs.csv']
# List of max steps and epochs trained
max_steps = [81, 150, 400, 729] * 3
epochs_trained = [50] * 4 + [10] * 4 + [100] * 4


for file_name, max_steps, epochs_trained in zip(file_names, max_steps, epochs_trained):
    # Load data from CSV without header
    df = pd.read_csv(file_name, header=None)

    # Assign column names
    df.columns = ['empty_cells', 'accuracy', 'avg_time', 'avg_steps']

    # Set the aesthetic style of the plots
    sns.set_theme(style="whitegrid")

    # Integers values on x axis

    # Plot 1: Accuracy vs Empty Cells
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x="empty_cells", y="accuracy", marker="o")
    plt.title(f'Accuracy vs. Number of Empty Cells for Max Steps set to {max_steps} and {epochs_trained} epochs trained')
    plt.xlabel('Number of Empty Cells')
    plt.ylabel('Accuracy')
    plt.xticks(np.arange(min(df["empty_cells"]), max(df["empty_cells"])+1, 1.0))
    plt.show()

    # Plot 2: Average Time vs Empty Cells
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x="empty_cells", y="avg_time", marker="o", color='orange')
    plt.title(f'Accuracy vs. Number of Empty Cells for Max Steps set to {max_steps} and {epochs_trained} epochs trained')
    plt.xlabel('Number of Empty Cells')
    plt.ylabel('Average Time (seconds)')
    plt.xticks(np.arange(min(df["empty_cells"]), max(df["empty_cells"])+1, 1.0))
    plt.show()

    # Plot 3: Average Steps vs Empty Cells
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x="empty_cells", y="avg_steps", marker="o", color='green')
    plt.title(f'Accuracy vs. Number of Empty Cells for Max Steps set to {max_steps} and {epochs_trained} epochs trained')
    plt.xlabel('Number of Empty Cells')
    plt.ylabel('Average Number of Steps')
    plt.xticks(np.arange(min(df["empty_cells"]), max(df["empty_cells"])+1, 1.0))
    plt.show()
```
