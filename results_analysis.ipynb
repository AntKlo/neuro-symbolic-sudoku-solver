```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# List of file names
file_names = ['results_max_steps_81.csv', 'results_max_steps_150.csv', 'results_max_steps_400.csv', 'results_max_steps_729.csv',
              'results_max_steps_81_10epochs.csv', 'results_max_steps_150_10epochs.csv', 'results_max_steps_400_10epochs.csv', 'results_max_steps_729_10epochs.csv',
              'results_max_steps_81_100epochs.csv', 'results_max_steps_150_100epochs.csv', 'results_max_steps_400_100epochs.csv', 'results_max_steps_729_100epochs.csv']
# List of max steps and epochs trained
max_steps = [81, 150, 400, 729] * 3
epochs_trained = [50] * 4 + [10] * 4 + [100] * 4


for file_name, max_step, epoch_trained in zip(file_names, max_steps, epochs_trained):
    # Load data from CSV without header
    df = pd.read_csv(file_name, header=None)

    # Assign column names
    df.columns = ['empty_cells', 'accuracy', 'avg_time', 'avg_steps']

    # Set the aesthetic style of the plots
    sns.set_theme(style="whitegrid")

    # Integers values on x axis

    # Plot 1: Accuracy vs Empty Cells
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x="empty_cells", y="accuracy", marker="o")
    plt.title(f'Accuracy vs. Number of Empty Cells for Max Steps set to {max_step} and {epoch_trained} epochs trained')
    plt.xlabel('Number of Empty Cells')
    plt.ylabel('Accuracy')
    plt.xticks(np.arange(min(df["empty_cells"]), max(df["empty_cells"])+1, 1.0))
    plt.show()

    # Plot 2: Average Time vs Empty Cells
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x="empty_cells", y="avg_time", marker="o", color='orange')
    plt.title(f'Avg time vs. Number of Empty Cells for Max Steps set to {max_step} and {epoch_trained} epochs trained')
    plt.xlabel('Number of Empty Cells')
    plt.ylabel('Average Time (seconds)')
    plt.xticks(np.arange(min(df["empty_cells"]), max(df["empty_cells"])+1, 1.0))
    plt.show()

    # Plot 3: Average Steps vs Empty Cells
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x="empty_cells", y="avg_steps", marker="o", color='green')
    plt.title(f'Avg steps vs. Number of Empty Cells for Max Steps set to {max_step} and {epoch_trained} epochs trained')
    plt.xlabel('Number of Empty Cells')
    plt.ylabel('Average Number of Steps')
    plt.xticks(np.arange(min(df["empty_cells"]), max(df["empty_cells"])+1, 1.0))
    plt.show()
```

```python
# Create a figure with 4 rows and 3 columns of subplots
fig, axes = plt.subplots(4, 3, figsize=(18, 24))
fig.suptitle('Aggregated Results for Different Max Steps and Epochs Trained', fontsize=16)
i = 0
for file_name, max_step, epoch_trained in zip(file_names, max_steps, epochs_trained):
    # Load data from CSV without header
    df = pd.read_csv(file_name, header=None)
    # Assign column names
    df.columns = ['empty_cells', 'accuracy', 'avg_time', 'avg_steps']

    # Plot Accuracy vs Empty Cells in the grid
    sns.lineplot(data=df, x="empty_cells", y="accuracy", marker="o", ax=axes[i//3, i%3])
    axes[i//3, i%3].set_title(f'Max Steps {max_step}, {epoch_trained} epochs')
    axes[i//3, i%3].set_xlabel('Empty Cells')
    axes[i//3, i%3].set_ylabel('Accuracy')
    axes[i//3, i%3].set_xticks(np.arange(min(df["empty_cells"]), max(df["empty_cells"])+1, 1.0))

    # Plot Average Time vs Empty Cells in the grid
    sns.lineplot(data=df, x="empty_cells", y="avg_time", marker="o", color='orange', ax=axes[i//3, i%3])
    axes[i//3, i%3].set_ylabel('Avg Time (seconds)')

    # Plot Average Steps vs Empty Cells in the grid
    sns.lineplot(data=df, x="empty_cells", y="avg_steps", marker="o", color='green', ax=axes[i//3, i%3])
    axes[i//3, i%3].set_ylabel('Avg Steps')

    i += 1

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
